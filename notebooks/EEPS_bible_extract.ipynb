```python
import json, os
AKJV = "../../DH/Early-Modern-Sermons/assets/Bibles/Bible-kjv-1611"

book_to_abbrev = {v:k for k,v in standard_books.items()}
book_to_abbrev.update({"Song of Solomon": "SNG", "Wisdom of Solomon":"WIS",
 "Letter of Jeremiah":"LETTER OF JEREMIAH", 
 "Prayer of Azariah": "S3Y", "Prayer of Manasseh":"MAN"})
standard_books.update({"LETTER OF JEREMIAH":"Letter of Jeremiah"})
book_parts.update({"LETTER OF JEREMIAH":"Apocrypha"})

bible = []
bible_dict = {}
version = "AKJV"
for fp in os.listdir(AKJV):
    if fp in ["README.md","Books.json","Books_chapter_count.json"]: continue 
    if "json" not in fp: continue 

    with open(f"{AKJV}/{fp}",'r') as file:
        book_dict = json.load(file)
        book_name = fp.split(".json")[0]
        book_abbrev = book_to_abbrev[book_name]
        book = standard_books[book_abbrev]

        part = book_parts[book_abbrev]

        for chapter in book_dict["chapters"]: 
            c_no = str(chapter['chapter'])
            for verse_info in chapter['verses']:
                v_no = str(verse_info["verse"])
                text = verse_info["text"]
                key = f"{book} {c_no}.{v_no} ({version})"
                bible.append({
                    "id": key,
                    "version": version,
                    "part": part, 
                    "book":book,
                    "chapter":c_no,
                    "verse":v_no,
                    "text":text
                })
                bible_dict[f"{book} {c_no}.{v_no}"] = text
output(version,bible,bible_dict)
```

```python
import requests 
from tqdm import tqdm 
from bs4 import BeautifulSoup

book_to_abbrev = {v:k for k,v in standard_books.items()}
version = "ODRV"

bible = []
for item in webpages: 
    if "exodus" not in item: continue 
    chap = 1 
    book = item
    if book in standardize: 
        book = standardize[book]
    elif "old" in book: 
        book = book.split("/")[1].capitalize()
    else: 
        book = book.capitalize()
    book_abbrev = book_to_abbrev[book]
    part = book_parts[book_abbrev]        

    webpage =  f'https://originaldouayrheims.com/{item}'

    content = requests.get(webpage).text
    chapters = re.findall(r"(\d+)</a>", str(content))
    last_ch = 1
    if len(chapters) > 0: 
        last_ch = max([int(x) for x in chapters])
    print(book)
    for c_no in tqdm(range(1,last_ch+1)):
        # if c_no <= 128: continue 
        if c_no > 1:
            webpage = f'https://originaldouayrheims.com/{webpages[item]+str(c_no)}'
            content = requests.get(webpage).text
        verses = re.findall(r'<b>\s*(\d+)[\.]*[\s]*</b>(?!</a>)(.*?)(?=<b>\s*\d+[\.]*[\s]*</b>|<span id="Annotations">|<style media="screen">|<nav class="Old">|<nav class="flexbox-item Old">|$)',str(content),re.DOTALL)
        c_no = str(c_no)
        for v_no, text in verses:
            text = re.sub(r"<i>|</i>|<p>|</p>|<br>|<r>","",text)
            text = re.sub(r'[\n\r]'," ",text)
            text = re.sub(r'\s+',' ',text).strip()
            pattern2 = r'<b class="tooltip[PR]*">.*?<span class="tooltip2">.*?</span>'
            text = re.sub(pattern2,"",text,re.DOTALL)
            pattern3 = r'<span class= ".*?">|<span id=".*?">|</span>'
            text = re.sub(pattern3,"",text)
            text = re.sub(r"<span>|</ul>|</b>","",text)
            text = re.sub(r'\s+',' ',text).strip()  
            key = f"{book} {c_no}.{v_no} ({version})"
            bible.append({
                "id": key,
                "version": version,
                "part": part, 
                "book":book,
                "chapter":c_no,
                "verse":v_no,
                "text":text
            })

    partial_output(version,bible)
```
