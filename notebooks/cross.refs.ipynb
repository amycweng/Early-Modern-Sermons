```python
import re,json
chap_lengths = {}
with open(f"../Early-Modern-Sermons/assets/Bibles/ESV.json",'r') as file:
  ESV = json.load(file)

for v_id in ESV: 
  BC = re.findall(f"([\w\s]+ \d+).",v_id)[0]
  if BC not in chap_lengths: 
    chap_lengths[BC] = 0 
  chap_lengths[BC] += 1 
  
def expand_verse_range(input):
    match = re.findall(r'(.+?) (\d+)\.(\d+)-(.+?) (\d+)\.(\d+)', input)[0]
    book1, ch1, v1, book2, ch2, v2 = match
    ch1, v1, ch2, v2 = map(int, [ch1, v1, ch2, v2])
    verses = []
    if ch1 == ch2 and book1==book2:
        for v in range(v1, v2 + 1):
            verses.append(f"{book1} {ch1}.{v}")
        return verses
    
    for v in range(v1, chap_lengths[f"{book1} {ch1}"] + 1):
        verses.append(f"{book1} {ch1}.{v}")
    for ch in range(ch1 + 1, ch2):
        for v in range(1, chap_lengths[f"{book2} {ch1}"] + 1):
            verses.append(f"{book2} {ch}.{v}")
    for v in range(1, v2 + 1):
        verses.append(f"{book2} {ch2}.{v}")
    return verses
```

```python
CR_dict = {}

def correct_id(item): 
    B = item.split(".")[0]
    return re.sub(f"{B}.",f"{corrected_books[B]} ",item)
    
for line in CR[1:]: 
    line = line.split("\t")
    u,vlist = line[0], line[1].split("-")
    u = correct_id(u)
    if u not in CR_dict: 
        CR_dict[u] = []
    for idx, v in enumerate(vlist): 
        vlist[idx] = correct_id(v)

    if len(vlist) == 1: 
        CR_dict[u].append(vlist[0]) 
    else: 
        CR_dict[u].extend(expand_verse_range("-".join(vlist)))
```
